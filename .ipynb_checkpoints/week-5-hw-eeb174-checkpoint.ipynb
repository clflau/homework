### 2. Which family has the most species represented?

This can be achieved by writing a bash script to use a for loop to cycle through all the list of families, and count the number of entries for each of those familes, and then sort for the highest number

```
#!/bin/bash

list_of_fam=$(tail -n +4 formatted_nature.csv | cut -d "," -f 3 | sort | uniq)

for fam in $list_of_fam
    do
        number=$(tail -n +4 formatted_nature.csv | cut -d "," -f 3 | sort | grep -c ^$fam)
        output+=$(printf "\n$fam, $number\n")
done

echo -e "$output\n" | sort -n -k 2 | tail
```
Trochilidae, 102 species

### 3. How many families have only a single species in them?

## Write a function that takes two arguments  a protein sequence and an amino acid residue code  and returns the percentage of the protein that the amino acid makes up.


````python
def aaPercent(protein, residue):
    percent = (protein.upper().count(residue.upper()) / len(protein))*100
    return percent
````

````python
assert aaPercent("MSRSLLLRFLLFLLLLPPLP", "M") == 5
````

````python
assert aaPercent("MSRSLLLRFLLFLLLLPPLP", "r") == 10
````

````python
assert aaPercent("msrslllrfllfllllpplp", "L") == 50
````

````python
assert aaPercent("MSRSLLLRFLLFLLLLPPLP", "Y") == 0
````

## Modify the function from part one so that it accepts a list of amino acid residues rather than a single one. If no list is given, the function should return the percentage of hydrophobic amino acid residues (A, I, L, M, F, W, Y and V).

````python
def aaPercent(protein, residue = ["A", "I", "L", "M", "F", "W", "Y", "V"]):
    PROTEIN = protein.upper()
    residuecount = []
    ticker = 0
    while ticker < len(residue) :
        residuecount.append(PROTEIN.count(residue[ticker]))
        ticker = ticker + 1
    allcount = sum(residuecount)
    percent = (allcount / len(PROTEIN)) * 100
    return int(percent)
````

````python
assert aaPercent("MSRSLLLRFLLFLLLLPPLP", ["M"]) == 5
assert aaPercent("MSRSLLLRFLLFLLLLPPLP", ['M', 'L']) == 55
assert aaPercent("MSRSLLLRFLLFLLLLPPLP", ['F', 'S', 'L']) == 70
assert aaPercent("MSRSLLLRFLLFLLLLPPLP") == 65
````
